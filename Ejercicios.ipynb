```python
rng_1 = np.random.default_rng(seed=1)

muestras_exponenciales = rng_1.exponential(scale=3, size=1000)
muestras_exponenciales
```

```python
df_exp = pd.DataFrame()

for i in range(1,101):
  muestras_exponenciales = rng_1.exponential(scale=3, size=1000)
  col = f"muestras_{i}"
  df_exp[col] = muestras_exponenciales

df_media = pd.DataFrame(df_exp.mean(), columns=["media de las muestras"])

df_media
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df_media, bins=32, kde=True)

plt.xlabel("Media de las muestras")
plt.ylabel("Frecuencia")
plt.title("Distribución de las medias de las muestras")
plt.show()
```
