```r
library(dplyr)
```

```r
data <- data.frame(
  name = c("Alice", "Bob", "Charlie"),
  age = c(25, 30, 35),
  score = c(90, 85, 88)
)

# Use dplyr to filter and select data
filtered_data <- data %>%
  filter(age > 25) %>%
  select(name, score)

# Print the filtered data
print(filtered_data)
```
